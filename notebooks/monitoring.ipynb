# Monitoring 

### Importing the required the libraries

Using Avidently.ai for monitoring the Model

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from evidently import ColumnMapping
from evidently.report import Report
from evidently.test_suite import TestSuite
from evidently.metric_preset import ClassificationPreset
from evidently.test_preset import DataQualityTestPreset
from evidently.test_preset import DataStabilityTestPreset
from evidently.metric_preset import TargetDriftPreset
from evidently.test_preset import DataDriftTestPreset
from evidently.test_preset import RegressionTestPreset
from evidently.test_preset import DataDriftTestPreset
from evidently.test_preset import NoTargetPerformanceTestPreset
from evidently.test_preset import MulticlassClassificationTestPreset
from evidently.test_preset import BinaryClassificationTopKTestPreset
from evidently.test_preset import BinaryClassificationTestPreset
print("Libraries Imported Succesfully!!")
```

### Importing the train and test data set 

```python
ref_data = pd.read_csv("preprocessed_data.csv")
cur_data = pd.read_csv("test_data.csv")
```

```python
ref_data.head()
```

```python
cur_data.head()
```

```python
ref_data['stroke'].value_counts()
```

```python
cur_data['stroke'].value_counts()
```

```python
ref_X = ref_data.drop(["stroke"], axis=1)
ref_y = ref_data["stroke"]

cur_X = cur_data.drop(["stroke"], axis=1)
cur_y = cur_data["stroke"]
```

```python
ref_X_train, ref_X_test, ref_y_train, ref_y_test = train_test_split(ref_X, ref_y, test_size=0.4, random_state=60)
```

```python
ref_y_test
```

```python
cur_X.head()
```

```python
ref_y.value_counts()
```

```python
rf = RandomForestClassifier()
rf.fit(ref_X_train.values, ref_y_train.values)
```

```python
ref_pred = rf .predict(ref_X_test)
ref_pred = pd.DataFrame(ref_pred, columns=["Prediction"])
cur_pred = rf.predict(cur_X)
cur_pred = pd.DataFrame(cur_pred, columns=["Prediction"])
```

### Joining Prediction Column into the dataframe

```python
# Merging on train data
ref_X_test.reset_index(inplace=True, drop=True)
ref_y_test.reset_index(inplace=True, drop=True)
ref_merged = pd.concat([ref_X_test, ref_y_test], axis=1)
ref_merged = pd.concat([ref_merged, ref_pred], axis=1)
ref_merged
```

```python
# Merging in test data
cur_X.reset_index(inplace=True, drop=True)
cur_y.reset_index(inplace=True, drop=True)
cur_merged = pd.concat([cur_X, cur_y], axis=1)
cur_merged = pd.concat([cur_merged, cur_pred], axis=1)
cur_merged
```

### Using Column mapping of Avidently.ai

```python
cm = ColumnMapping()
cm.target = "stroke"
cm.prediction = "Prediction"
cm.target_names = ["No stroke", "stroke"]
```

```python
from evidently.metric_preset import DataQualityPreset
data_quality_report = Report(metrics=[
    DataQualityPreset(),
])

data_quality_report.run(reference_data=ref_merged, current_data=cur_merged)
data_quality_report.save_html("reports/dataqualityreport.html")

```

```python
num_target_drift_report = Report(metrics=[
    TargetDriftPreset(num_stattest='ks', cat_stattest='psi'),
])

num_target_drift_report.run(reference_data=ref_merged, current_data=cur_merged)
num_target_drift_report.save_html("reports/num_target_drift_report.html")
```

```python
classification_performance_report = Report(metrics=[
    ClassificationPreset(),
])
classification_performance_report.run(reference_data=ref_merged, current_data=cur_merged, column_mapping = cm)
classification_performance_report.save_html("reports/classification_performance_report1.html")
```

```python
ref_probas = pd.DataFrame(rf.predict_proba(ref_X_test))
ref_probas.columns = ["No stroke", "stroke"]
cur_probas = pd.DataFrame(rf.predict_proba(cur_X))
cur_probas.columns = ["No stroke", "stroke"]
```

```python
ref_probas
```

```python
ref_probas.value_counts()
```

```python
ref_X_test.reset_index(inplace=True, drop=True)
ref_X_test['Attrition'] = ['No Stroke' if x == 0 else 'Stroke' for x in ref_y_test]
ref_merged = pd.concat([ref_X_test, ref_probas], axis = 1)

cur_X.reset_index(inplace=True, drop=True)
cur_X['Attrition'] = ['Stroke' if x == 0 else 'Stroke' for x in cur_y]
cur_merged = pd.concat([cur_X, cur_probas], axis = 1)
```

```python
ref_merged
```

```python
cm  = ColumnMapping()
cm.target = 'Attrition'
cm.prediction = ['No stroke', 'stroke']
cm.pos_label = 'No stroke'
```

```python
classification_performance_report = Report(metrics=[
    ClassificationPreset(),
])
classification_performance_report.run(reference_data=ref_merged, current_data=cur_merged, column_mapping = cm)
classification_performance_report.save_html("Reports/classification_performance_report2.html")
```

```python
data_quality = TestSuite(tests=[
    DataQualityTestPreset(),
])

data_quality.run(reference_data=ref_merged, current_data=cur_merged)
data_quality.save_html("reports/data_quality.html")
```

```python
data_drift = TestSuite(tests=[
    DataDriftTestPreset(stattest='psi'),
])

data_drift.run(reference_data=ref_merged, current_data=cur_merged)
data_drift.save_html("reports/data_drift.html")
```
